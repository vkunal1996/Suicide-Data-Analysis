## Statewise- YearWise Analysis 

```python


"""pd=np.where(a[...,0]=="Punjab")
print(a[pd])
print()
year=np.where(np.logical_and(a[...,0]=="Punjab",a[...,1]=="2008"))
print(a[year])"""
states=np.array(a[1:,...],dtype=str)

def findsuicides(state,year):
    data=states[states[:,0]==state]
    deaths=np.array(data[:,6],dtype=int)
    b=np.where(data[:,1]==year)
    #numberofDeaths=np.array(b[:,6] , dtype=int)
    print("Total Suicides in ",state," in ",year," are ",np.sum(deaths[b]))
```

```python
s=np.array(a[1:,0])
y=input("Enter Year : ")
statesset=set(s)
stateslist=[value for value in statesset]
stateslist.sort()
for i in stateslist:
    findsuicides(i,y)
```

### Statewise - Analysis

```python
def find(state):
    data=states[states[:,0]==state]
    deaths=np.array(data[:,6],dtype=int)
    print("Total Suicides in {} are {}".format(state,np.sum(deaths)))
```

```python
for i in stateslist:
    find(i)
```

## YearWise Analysis

```python
def yearwise(y):
    year=np.array(a[np.where(a[:,1]==y)])
    deaths=np.array(year[:,6],dtype=int)
    print("Suicides in the Year ",y," ", np.sum(deaths))
```

## GenderWise Analysis

```python
female=a[a[:,4]=="Female"]
male=a[a[:,4]=="Male"]
male_commit_suicide=np.array(male[:,6],dtype=int)
female_commit_suicide=np.array(female[:,6],dtype=int)
print("Male Suicide Count is {} and Female Suicide Count is {}".format(np.sum(male_commit_suicide),np.sum(female_commit_suicide)))

```

#### Statewise - Gender Wise analysis

```python
def stategender(state):
    m=np.where(np.logical_and(a[...,4]=="Male",a[...,0]==state))
    male=a[m]
    male_commit_suicide=np.array(male[...,6],dtype=int)
    m=np.where(np.logical_and(a[...,4]=="Female",a[...,0]==state))
    female=a[m]
    female_commit_suicide=np.array(female[...,6],dtype=int)
    
    print("In {} \nMale Suicide : {} \tFemale Suicide :{}".format(state,np.sum(male_commit_suicide),np.sum(female_commit_suicide)))
```

```python
for i in stateslist:
    stategender(i)
```

#### Genderwise -YearWise analysis

```python
def yeargender(year):
    y=np.where(np.logical_and(a[...,4]=="Male",a[...,1]==year))
    male=a[y]
    male_commit_suicide=np.array(male[...,6],dtype=int)
    y=np.where(np.logical_and(a[...,4]=="Female",a[...,1]==year))
    female=a[y]
    female_commit_suicide=np.array(female[...,6],dtype=int)
    
    print("In {} \nMale Suicide : {} \tFemale Suicide :{}".format(year,np.sum(male_commit_suicide),np.sum(female_commit_suicide)))
```

```python
for i in yearlist:
    yeargender(i)
```

#### Statewise-yearwise-genderwise analysis

```python
def stateyeargender(state,year):
    temp=np.where(np.logical_and(np.logical_and(a[...,0]==state,a[...,1]==year),a[...,4]=="Male"))
    male=a[temp]
    male_commit_suicide=np.array(male[...,6],dtype=int)
    temp=np.where(np.logical_and(np.logical_and(a[...,0]==state,a[...,1]==year),a[...,4]=="Female"))
    female=a[temp]
    female_commit_suicide=np.array(female[...,6],dtype=int)
    
    print("In {} \nMale Suicide : {} \tFemale Suicide : {}".format(year,np.sum(male_commit_suicide),np.sum(female_commit_suicide)))
```

```python
for i in stateslist:
    print("In {}".format(i))
    print()
    for j in yearlist:
        stateyeargender(i,j)
    print()
    print()
```
